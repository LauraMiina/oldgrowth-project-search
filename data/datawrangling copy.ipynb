```python
df3 = pd.read_csv('projectlisting.csv')
df3.head()
```

```python
df3 = df3.rename(columns={'Unnamed: 1': 'name_of_project'})
df3
```

```python
#show the type of data in the columns
df3.dtypes
```

```python
df3['project'] = df3['project'].astype(str)
```

```python
df3['name_of_project'] = df3['project'].str.extract(r'(.*)\(')
df3
```

```python
df3['project'] = df3['project'].str.extract(r'\((.*)\)')
df3.head()
```

```python
df3.to_csv('projectlisting.csv', index=False)
```
