# Cleaning data and converting the project listing from a csv to a geojson

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
df = pd.read_csv('projectlisting.csv', encoding='latin1')
df
```

```python
df.dtypes
```

```python
df.rename(columns={'Acres subject to any logging':'acres_to_log'}, inplace=True)
```

```python
df.to_csv('projectlisting_7_29.csv', index=False)
```

```python
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.Longitude, df.Latitude))
```

```python
gdf.dtypes
```

```python
gdf.to_file('projectlisting_7_29.geojson', driver='GeoJSON')
```
