```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# open projectlisting.xlsx as a pandas dataframe
df = pd.read_excel('projectlisting.xlsx')
df
```

```python
import pandas as pd
import re

# Function to extract detailed parts from the original content
def extract_details(text):
    project = re.search(r'^(.*?\(.*?\)):', text)
    description = re.search(r'^(.*?\(.*?\)): (.*?)( Update:| Local contact:|$)', text)
    update = re.search(r'Update: (.*?)( Local contact:|$)', text)
    contact = re.search(r'Local contact: (.*)', text)
    
    project = project.group(1) if project else ''
    description = description.group(2) if description else ''
    update = update.group(1) if update else ''
    contact = contact.group(1) if contact else ''
    
    other = text
    parts_to_remove = [project, description, 'Update: ' + update if update else '', 'Local contact: ' + contact if contact else '']
    for part in parts_to_remove:
        if part:
            other = other.replace(part, '')
    other = other.strip()
    
    return pd.Series([project, description, update, contact, other])

# Apply the function to extract details
df_details = df.apply(lambda x: extract_details(x[0]), axis=1)

# Assign column names
df_details.columns = ['Project', 'Description', 'Update', 'Contact', 'Other']

# Fill NaN values with empty strings
df_details.fillna('', inplace=True)

```

```python
# create new columns and leave them empty: Litigation, Paused or ongoing, Sent to review, Amount of old growth
df_details['Litigation'] = ''
df_details['Paused or ongoing'] = ''
df_details['Sent to review'] = ''
df_details['Amount of old growth'] = ''

```
